# 파이썬으로 CSV 파일 출력하기

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv') as f:
  print(f.readline())
```

위의 오류는 UTF-8 코덱이 어떤 바이트를 읽을 수 없음

파이썬의 open() 함수는 기본적으로 텍스트 파일이 UTF-8 형식으로 저장되어 있다고 가정

하지만 한글 텍스트는 EUC-KR을 자주 사용

#### 따라서 chardet.detect() 함수를 이용하여 문자 인코딩 방식을 알아냄
mode = 'rb' (binary 읽기모드)

```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv',mode = 'rb') as f:
  d = f.readline()
print(chardet.detect(d))
```

#### 따라서 encoding 매개변수로 인코딩 형식을 'EUC-KR'로 지정해야함

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv',encoding='EUC-KR') as f:
  print(f.readline())
  print(f.readline())
```

# 판다스(pandas)
### CSV 파일은 한 줄씩 읽어서 데이터 분석하기 어렵기에, 판다스 패키지 사용
1. 데이터 프레임
: 행과 열로 구성된 데이터 구조
2. 시리즈
: 1차원 배열과 비슷 (시리즈에 담긴 데이터는 모두 동일한 종류)

## CSV 파일을 데이터 프레임으로 읽기
### read_csv() 함수
: 판다스에서 CSV 파일을 읽음

DtypeWarning: Columns (5,6,9) have mixed types. Specify dtype option on import or set low_memory=False.

: 판다스는 데이터 타입을 자동으로 설정하는데, 이 설정과 실제 데이터 값이 다른 자료형일때 발생하는 경고

즉 이 파일은 5,6,9 열 때문에 오류가 발생

--> low_memory 매개변수를 False로 지정하여 파일을 나누어 읽지 않고 한 번에 읽어야함.

```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv',encoding='EUC-KR')
```

```python
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv',encoding='EUC-KR', low_memory = False)
```

## head() 메서드
: 데이터프레임의 처음 다섯 개 행을 확인할 수 있음  
이때, 주어진 데이터 이외의 첫번째 열에서 숫자가 나오는데 판다스가 행마다 0부터 시작하는 인덱스 번호를 자동으로 붙여준 것.  

**_첫 행이 열 이름이 아니라면_**   
read_csv() 함수를 호출할 때, header 매개변수를 None으로 지정해주어 테이터 첫 행에 열 이름이 없다는 것을 알리고,   
names 매개변수에 열 이름 리스트를 따로 전달해줌 -> 열 이름이 중복되면 안됨

결과에 원래 CSV 파일에는 없던 'Unnamed:13' 열이 생김   
CSV 파일에는 각 라인의 끝에 콤마(,)가 있음  
판다스는 이 콤마를 보고 마지막에 하나의 열이 더 있다고 판단해서 나온것

## 데이터프레임을 CSV 파일로 저장하기 : to_csv() 메서드
: to_csv() 메서드는 UTF-8 형식으로 저장하기 때문에 나중에 open() 함수로 파일을 읽을 때 따로 encoding 매개변수를 사용하지 않아도 됨.

```python
df.to_csv('ns_202104.csv')
```

```python
with open('ns_202104.csv') as f:
  for i in range(3):
    print(f.readline(), end='')
```

```python
ns_df = pd.read_csv('ns_202104.csv',low_memory=False)
ns_df.head()
```

위의 코드는 index가 같이 저장되었음

인덱스 열과 Unnamed:0 이 중복됨   
따라서 CSV 파일에 인덱스가 이미 있다는 것을 알려주려면 **index_col** 매개변수를 사용해야함

```python
ns_df=pd.read_csv('ns_202104.csv',index_col=0,low_memory=False)
ns_df.head()
```

다른 방법으로 애초에 데이터프레임을 CSV 파일로 저장할 때, 인덱스를 빼고 저장할 수도 있음.

```python
df.to_csv('ns_202104.csv',index=False)
```
